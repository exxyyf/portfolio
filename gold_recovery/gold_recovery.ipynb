# Восстановление золота из руды

**Описание проекта**

Целью данного проекта является подготовка модели машинного обучения для компании "Цифра". Задачей модели является **предсказание коэффициента восстановления золота из золотосодержащей руды.** По заданию нужно спрогнозировать эффективность обогащения как **чернового концентрата** `rougher.output.recovery`, так и **финального концентрата** `final.output.recovery`.

В нашем распоряжении находятся три выборки (обучающая, тестовая и исходная) сырых данных с множеством признаков, таких как размер гранул сырья, концентрация металлов и прочие. Необходимо изучить данные, обработать их, выделить признаки, нужные для обучения модели.

**Приняты следующие обозначения признаков:**

`[этап].[тип_параметра].[название_параметра]`

Возможные значения для блока `[этап]`:

- rougher — флотация
- primary_cleaner — первичная очистка
- secondary_cleaner — вторичная очистка
- final — финальные характеристики

Возможные значения для блока `[тип_параметра]`:

- input — параметры сырья
- output — параметры продукта
- state — параметры, характеризующие текущее состояние этапа
- calculation — расчётные характеристики

**Метрикой качества модели** будет служить симметричное среднее абсолютное процентное отклонение (sMAPE), вычисляемое по следующей формуле:

$$sMAPE = \frac{1}{N} \sum_{i=1}^{N} \frac{|y_i - \hat{y_i}|}{(|{y_i}| + |\hat{y_i}|) / 2} * 100\% $$

Эта формула одинаково (симметрично) учитывает как целевой признак, так и предсказания модели. Измеряется sMAPE в относительных величинах.

Так как требуется прогноз двух величин, итоговая метрика качества будет рассчитана по формуле:

$$sMAPE(result) = 25\% * sMAPE(rougher) + 75\% * sMAPE(final)$$

<h3>Проверка расчета эффективности обогащения</h3>

```python
def recovery_mae(concentrate, feed, tail, recovery):
    test_recovery = ((concentrate * (feed - tail)) / (feed * (concentrate - tail))) * 100
    recovery_mae = mean_absolute_error(recovery, test_recovery)
    
    return recovery_mae
```

```python
i = 0

mae = recovery_mae(df_list[i]['rougher.output.concentrate_au'],
                     df_list[i]['rougher.input.feed_au'],
                     df_list[i]['rougher.output.tail_au'],
                     df_list[i]['rougher.output.recovery'])

print(f'\nMAE для эффективности обогащения золота на обучающей выборке составляет: {mae} или {mae:.15f}\n')
```

<h4>Масштабирование данных</h4>

Так как мы увидели, что большая часть признаков имеет нормальное распределение, то будем использовать StandartScaler для масштабирования. Fit будем делать только на тренировочной выборке, а трансформировать уже будем обе.

```python
def scale(features_train, features_valid):
    
    """
    scales target and features dfs
    
    returns tuple with 2 dfs
    """
    
    scaler = StandardScaler()
    scaler.fit(features_train)
    
    col_train = features_train.columns
    col_valid = features_valid.columns
    
    index_train = features_train.index
    index_valid = features_valid.index
    
    features_train = scaler.transform(features_train)
    features_valid = scaler.transform(features_valid)
    
    features_train = pd.DataFrame(features_train, columns = col_train, index=index_train)
    features_valid = pd.DataFrame(features_valid, columns = col_valid, index=index_valid)
    
    return features_train, features_valid
```

```python
features_train, features_test = scale(features_train, features_test)
```

## Модель

<h3>Написание функций для sMAPE</h3>

Для начала напишем функцию для расчета интересующей нас метрики качества моделей - sMAPE:

sMAPE для каждой из моделей:

$$sMAPE = \frac{1}{N} \sum_{i=1}^{N} \frac{|y_i - \hat{y_i}|}{(|{y_i}| + |\hat{y_i}|) / 2} * 100\% $$

Итоговая sMAPE:

$$sMAPE(result) = 25\% * sMAPE(rougher) + 75\% * sMAPE(final)$$

```python
def smape(target, predict):
    """
    takes two arrays with target features and predictions,
    returns sMAPE
    """
    smape = (abs(target - predict) / ((abs(target) + abs(predict)) / 2)) * 100
    smape = smape.mean()
    return smape
```

```python
def final_smape(rougher, final):
    
    smape = 0.25*rougher + 0.75*final
    
    return smape
```

<h3>Обучение моделей</h3>

В качестве моделей мы будем обучать и тестировать линейную регрессию и случайный лес. Оценивать их будем кросс-валидацией.

```python
linear_model = LinearRegression()
scorer = make_scorer(smape)

linear_score = cross_val_score(linear_model, features_train, target_train_rougher, cv=5, scoring=scorer)

linear_score_final = sum(linear_score)/len(linear_score)

print(f'Средняя оценка качества модели линейной регрессии (rougher): {linear_score_final:.4f}%' )
```

```python
linear_model_fin = LinearRegression()

linear_score_fin = cross_val_score(linear_model_fin, features_train, target_train_final, cv=5, scoring=scorer)

linear_score_fin = sum(linear_score_fin)/len(linear_score_fin)

print(f'Средняя оценка качества модели линейной регрессии (final): {linear_score_fin:.4f}%' )
```

```python
def forest_maker(features, target):
    best_score = []
    best_trees = []
    best_depth = []

    for x in range(10, 51, 10):
        for y in range(1, 6):
            model = RandomForestRegressor(n_estimators=x, max_depth=y, random_state=1212)
            score = cross_val_score(model, features, target, cv=5, scoring=scorer)
            final_score = sum(score)/len(score)
            best_score.append(final_score)
            best_trees.append(x)
            best_depth.append(y)
            
    first = best_score[0]
    for i in range(len(best_score)):
        if best_score[i] < first:
            best = best_score[i]
            best_index = best_score.index(best)
        else:
            pass
    
    print(f'Лучшее значение sMAPE: {best_score[best_index]:.4f}%\nКоличество деревьев: {best_trees[best_index]}\nГлубина дерева: {best_depth[best_index]}')
```

```python
forest_maker(features_train, target_train_rougher)
```

```python
forest_maker(features_train, target_train_final)
```

Получается, что лучший результат, то есть наименьшее sMAPE был достигнут с помощью алгоритма случайного леса с 50 деревьями глубиной 5. Данное утверждение верно как для rougher, так и для final.

Для rougher sMAPE составила 6.3454%, final sMAPE составила 9.0104%.

<h3>Проверка моделей на вменяемость</h3>

Теперь проведем sanity check наших моделей с помощью константной модели. В качестве константной возьмем модель, предсказания которой будут равны среднему значению целевого признака.

```python
predictions_rougher = pd.Series(target_test_rougher.mean(), index=target_test_rougher.index)
predictions_final = pd.Series(target_test_final.mean(), index=target_test_final.index)

smape_r = smape(target_test_rougher, predictions_rougher)
smape_f = smape(target_test_final, predictions_final)

print(f'sMAPE для константной модели (rougher): {smape_r:.4f}%\n')
print(f'sMAPE для константной модели (final): {smape_f:.4f}%\n')

smape_const = final_smape(smape_r, smape_f)
print(f'Итоговое sMAPE для константной модели: {smape_const:.4f}%\n')
```

Итак, итоговое значение sMAPE меньше у константной модели, чем у моделей линейной регрессии. То есть делаем вывод о том, что использование моделей машинного обучения в данном случае не очень эффективно.
