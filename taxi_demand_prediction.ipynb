## Обучение

<h3>Создание признаков</h3>

Для обучения моделей нам необходимо создать признаки. 

В число этих признаков войдут:

- `календарные признаки`: месяц, день недели, час. Год брать не будем, так как у нас данные одного года;

- `"отстающие" значения` - возьмем отставание на 72 часа, то есть чтобы было оценено влияние на трое суток вперед;

- `скользящее среднее` с окном в 12 часов, так как оно достаточно хорошо аппроксимирует данные.

Напишем функцию для создания признаков:

```python
def make_features(data, max_lag, window, target):
    """
    takes df, target feature, lag quantity, rolling mean window size
    makes features for ml: month, dayofweek, hour, lags, rolling mean
    """
    data['month'] = data.index.month
    data['dayofweek'] = data.index.dayofweek
    data['hour'] = data.index.hour
    
    for lag in range(1, max_lag+1):
        data['lag_{}'.format(lag)] = data[target].shift(lag)
        
    data['rolling_mean'] = data[target].shift().rolling(window).mean()
```

<h3>Обучение моделей</h3>

В данном разделе проекта мы обучим следующие модели:

- линейная регрессия
- линейная регрессия с регуляризацией с помощью евклидового расстояния (Ridge)

В качестве метрики качества будем использовать среднеквадратическое отклонение - RMSE.

```python
def rmse(target, predictions):
    rmse = ((target - predictions)**2).mean()**0.5
    return rmse
```

```python
best_lr_rmse
```

Среднеквадратическое отклонение линейной регрессии на валидационной выборке составило 31. Порог по заданию - 48.

<h4>Ridge регрессия</h4>

В ридж модели будем изменять параметр `alpha`:

```python
best_alpha = None
best_model_r = None
best_r_rmse = 100

for a in [0.2, 0.3, 0.4, 0.6, 0.8, 1]:
    r_model = Ridge(alpha=a)
    r_model.fit(features_train, target_train)
    pred_ridge = r_model.predict(features_valid)
    rmse_r = rmse(target_valid, pred_ridge)
    
    if rmse_r < best_r_rmse:
        best_r_rmse = rmse_r
        best_alpha = a
        best_model_r = r_model
```

```python
best_alpha, best_r_rmse
```

Видим, что результат практически такой же, как у линейной регрессии. 

Можем сделать вывод о том, что в данном конкретном случае применение регуляризации для линейной регрессии не приносит результата.

<h4>Случайный лес</h4>

Для выбора лучшей модели случайного леса будем изменять параметры `max_depth` и `n_estimators`.

```python
b_est = None
b_depth = None
best_forest = None
best_f_rmse = 100

for est in range(50, 100, 10):
    for depth in range(1, 21, 2):
        f_model = RandomForestRegressor(max_depth=depth, n_estimators=est, random_state=RS)
        f_model.fit(features_train, target_train)
        pred_f = f_model.predict(features_valid)
        rmse_f = rmse(target_valid, pred_f)
        
        if rmse_f < best_f_rmse:
            best_f_rmse = rmse_f
            b_est = est
            b_depth = depth
            best_forest = f_model
```

```python
b_est, b_depth, best_f_rmse
```

Лучший результат среднеквадратического отклонения - 31 - показала модель с 80 деревьями в ансамбле и максимальной глубиной 17. 

<h4>Градиентный бустинг LightGBM</h4>

Для модели LightGBM будем изменять такие параметры, как скорость обучения `learning_rate` и максимальную глубину деревьев `max_depth`.

```python
best_rate = None
best_est = None
b_depth = None
best_lgb = None
best_lgb_rmse = 100

for rate in [0.2, 0.4, 0.5]:
    for est in range(50, 151, 50):
        for depth in range(1, 11, 2):
            model_lgb = ltb.LGBMRegressor(random_state=RS, max_depth=depth, learning_rate=rate, num_threads=8, n_estimators=est)
            model_lgb.fit(features_train, target_train)
            pred_lgb = model_lgb.predict(features_valid)
            rmse_lgb = rmse(target_valid, pred_lgb)

            if rmse_lgb < best_lgb_rmse:
                best_lgb_rmse = rmse_lgb
                b_depth = depth
                best_rate = rate
                best_lgb = model_lgb
                best_est = est
```

```python
best_rate, b_depth, best_est, best_lgb_rmse
```

Лучший результат - RMSE=29.9 - показала модель со скоростью обучения 0.2, максимальной глубиной деревьев равной 7, 100 деревьями. 

<h4>Градиентный бустинг CatBoost</h4>

Для данной модели будем изменять такие параметры, как глубина `depth` и количество итераций `iterations`.

```python
best_depth = None
best_iter = None
best_cat = None
best_cat_rmse = 100

for i in range(50, 151, 50):
    for depth in range(1, 11, 2):  
        model_cat = ctb.CatBoostRegressor(loss_function="RMSE", random_state = RS, verbose=False, iterations=i, depth=depth)
        model_cat.fit(features_train, target_train)
        pred_cat = model_cat.predict(features_valid)
        rmse_cat = rmse(target_valid, pred_cat)
        
        if rmse_cat < best_cat_rmse:
            best_cat_rmse = rmse_cat
            best_depth = depth
            best_iter = i
            best_cat = model_cat
```

```python
best_depth, best_iter, best_cat_rmse
```

Лучший результат показала модель CatBoost с глубиной 7 и 150 итерациями. RMSE=31.
