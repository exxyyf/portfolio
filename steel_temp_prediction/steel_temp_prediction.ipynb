# Предсказание температуры стали для оптимизации производства


**Цель проекта:**

Задачей данного проекта является создание модели машинного обучения, способной предсказать температуру стали, для металлургического комбината ООО "Так закаляем сталь". Предсказание должно помочь оптимизировать энергозатраты на этапе обработки стали. 


**Процесс производства** состоит из следующих итераций:

 1. Измерение температуры и химический анализ стали
 2. Повышение температуры

Итак, перед нами информация о структуре семи датафреймов. Видим, что названия стобцов не соответствуют стадартному для Python **snake_case**: встречаются как русские названия, так и английские. 

Также видим, что в колонках разное число наблюдений, есть пропущенные значения, встречаются разные типы данных.

Русские названия столбцов переведем на английский язык и заменим вручную, английские - с помощью функции, которую напишем ниже.

```python
def snake_case(column):
    """
    converts a CamelCase string to snake_case
    """
    snake_case = ''
    for i in range(len(column)):
        if (column[i].isupper()) & (i == 0):
            snake_case += column[i].lower()
        elif column[i].isupper():
            snake_case += "_" + column[i].lower()
        elif column[i] == ' ':
            snake_case += "_"
        else:
            snake_case += column[i]
    return snake_case
```

```python
for i in [1, 2, 5, 6]:
    column_list = list(map(snake_case, df_list[i].columns))
    df_list[i].columns = column_list
```

```python
def analize(df, bins, color, figsize, title):
    """
    takes DataFrame, bins count for histplot and color of plot, figsize and title
    prints head, info, describe, NaN plot, histplot
    """
    display(df.head())
    print('---------------------------------------------------------------\n')
    display(df.info())
    print('---------------------------------------------------------------\n')
    display(df.describe().style.format('{:.4f}'))
    print('---------------------------------------------------------------\n')
    
    if df.isna().any().sum() == 0:
        print("No missing values detected")
    else:
        data_na = df.isna().sum().sort_values(ascending=False)
        ax = data_na[data_na>0].plot(kind='bar', figsize=(17,3), grid=False, color=color)

        rects = ax.patches
        labels = data_na

        for rect, label in zip(rects, labels):
            height = rect.get_height()
            ax.text(
                rect.get_x() + rect.get_width() / 2, height + 5, label, ha="center", va="bottom", fontsize=8
            )

        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.spines['bottom'].set_visible(True)
        ax.spines['left'].set_visible(False)   

        ax.set(xlabel="\nСтолбцы с пропущенными значениями",
        ylabel="\nКол-во пропущенных значений")
        plt.xticks(rotation = 45, ha = 'right', fontsize=8)
        ax.set_title(f"График пропусков в {title}", fontsize=12)

        plt.show()
    
    print('---------------------------------------------------------------\n')
    print("Гистограмма распределения признаков")
    df.hist(figsize=figsize, bins=bins, color=color)
    plt.show()
```
